```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df_skill['gabung_fitur'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_sim_df = pd.DataFrame(
    cosine_sim,
    index = df_skill['title'] + ' (' + df_skill.index.astype(str) + ')',
    columns = df_skill['title'] + ' (' + df_skill.index.astype(str) + ')'
)
```

```python
def recommend_by_title(title, cosine_sim, df_skill, items, k=5):
    """
    title: str, judul (misal 'Python')
    cosine_sim: numpy array hasil cosine_similarity
    df_skill: DataFrame asli dengan kolom 'title'
    items: DataFrame subset berisi kolom yang ingin ditampilkan
    k: jumlah rekomendasi
    """
    # Cari judul yang mengandung kata kunci (case-insensitive)
    matches = df_skill[df_skill['title'].str.contains(title, case=False, na=False)]

    if matches.empty:
        raise ValueError(f"Tidak ada judul yang mengandung kata '{title}'.")

    # Ambil index pertama yang cocok
    idx = matches.index[0]
    index_title = matches.iloc[0]['title'] + ' (' + str(idx) + ')'

    if index_title not in cosine_sim_df.index:
        raise ValueError(f"Index gabungan '{index_title}' tidak ditemukan di cosine_sim_df.")

    # Ambil index di cosine_sim_df
    idx_in_cosine = cosine_sim_df.index.get_loc(index_title)

    # Hitung skor kemiripan
    sim_scores = list(enumerate(cosine_sim[idx_in_cosine]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Ambil k+1 teratas (skip diri sendiri)
    topk = sim_scores[1:k+1]
    rec_indices = [i for i, _ in topk]
    scores = [s for _, s in topk]

    # Ambil baris dari df_skill
    rec_titles = [cosine_sim_df.index[i] for i in rec_indices]
    rec_indices_df = [int(t.split('(')[-1].strip(')')) for t in rec_titles]

    rec_df = items.iloc[rec_indices_df].copy()
    rec_df['Similarity Score'] = scores
    return rec_df
```
